```python
from qiskit import QuantumCircuit 
```

```python
n = 4 
qc =QuantumCircuit(n,n-1)
qc.x(-1)

for i in range(n):
    qc.h(i)
# qc.barrier()
qc.barrier()
# building f(x) function with certain secret string
# when the bit in secret string is 1, add the cnot gate to 'turn on the query', also cnot gate act as the \oplus operation
# in this case, we already know that the secret bitstring is 101

qc.cx(0,-1),
# qc.cx(1,-1)
qc.cx(2,-1)
# qc.cx()
# qc.h(0), qc.h(2)
qc.barrier()
for i in range(n-1):
    qc.h(i)
for i in range(n-1):
    qc.measure(i,i)

```

```python
qc.draw('mpl', fold=-1)
```

```python
# shot-based measurement # 
from qiskit_aer import AerSimulator
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import Session, SamplerV2 as Sampler

aer_sim = AerSimulator()
pm = generate_preset_pass_manager(backend=aer_sim, optimization_level=3)
isa_qc = pm.run(qc)

shots = 10000
with Session(backend=aer_sim) as session:
    sampler = Sampler(aer_sim)
    result = sampler.run([isa_qc], shots= shots).result()
expval = []
counts = result[0].data.c.get_counts()
p_0, p_1 = 0, 0
for key in counts.keys():
    print('The secret bitstring is:',key)
```
